```python
# TASK 1
# importing pandas library
import pandas as pd 
```

```python
# Creating the Employee DataFrame
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]  # created dictionary to create employee dataframe using key and list values
}
df_employee = pd.DataFrame(employee_data) 
df_employee.to_csv('employee.csv', index=False)  # Saving DataFrame to a CSV file

# Displaying the dataframe
df_employee
```

```python
# Creating a dictionary for seniority levels
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],  # Employee IDs
    'Designation Level': [2, 2, 3, 2, 3]  # Designation level of each employee
}

# Creating a DataFrame from the dictionary
df_seniority = pd.DataFrame(seniority_data)

# Saving the DataFrame to a CSV file
df_seniority.to_csv('seniority.csv', index=False)

# Displaying the seniority DataFrame
df_seniority
```

```python
# Creating a dictionary for project details

project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7',
                'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],  # Some costs are missing
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed',
               'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']  # Status of each project
}

# Creating a DataFrame from the dictionary
df_project = pd.DataFrame(project_data)

# Saving the DataFrame to a CSV file
df_project.to_csv('project.csv', index=False)

# Displaying the project DataFrame
df_project
```

```python
# TASK 2 
# Import pandas to work with data
import pandas as pd

# Reading of the project.csv file
df_project = pd.read_csv('project.csv')

# Creating an empty list to store updated cost values
new_costs = []

# Setting up total and count to calculate average later
total = 0     # Will keep the sum of known costs
count = 0     # Will count how many known costs we have seen

# Loop through each value in the 'Cost' column
for cost in df_project['Cost']:
    
    # Checking if cost is missing (NaN)
    if pd.isna(cost):
        # If it's the first value and nothing is available
        if count == 0:
            average = 0
        else:
            average = total / count  # Calculating running average from known values
        
        # Replace missing value with running average
        new_costs.append(average)
    
    else:
        # If the cost is not missing, keeping the original value
        new_costs.append(cost)
        
        # Updating total and increase count for future averages
        total = total + cost
        count = count + 1

# Replacing the old 'Cost' column with the new filled values
df_project['Cost'] = new_costs

# Saving the updated data back to the CSV file
df_project.to_csv('project.csv', index=False)

# Displaying the updated DataFrame
df_project
```

```python
# TASK 3 
# Reading employee data
df_employee = pd.read_csv('employee.csv')

# Splitting 'Name' column into 'First Name' and 'Last Name'
name_split  = df_employee['Name'].str.split(pat=' ', n=1, expand=True)
df_employee['First Name'] = name_split[0]
df_employee['Last Name'] = name_split[1]

# Dropping the old 'Name' column
df_employee.drop('Name', axis=1, inplace=True)

# Saving updated DataFrame back to the same CSV file
df_employee.to_csv('employee.csv', index=False)

# Displaying the updated employee DataFrame
df_employee
```

```python
# TASK 4 

# Reloading all CSVs
df_employee = pd.read_csv('employee.csv')
df_project = pd.read_csv('project.csv')
df_seniority = pd.read_csv('seniority.csv')

# Merging employee with seniority on 'ID'
employee_with_seniority = pd.merge(df_employee, df_seniority, on='ID')

# Merging the result with project data
final_df = pd.merge(employee_with_seniority, df_project, on='ID')

# Displaying the final merged DataFrame
final_df
```

```python
# TASK 5 

# Adding 'Bonus' column - 5% of cost if status is 'Finished'
final_df['Bonus'] = final_df.apply(lambda row: 0.05*row['Cost'] if row['Status'] == 'Finished' else 0, axis=1)

# Displaying the DataFrame with bonus
final_df
```

```python
# TASK 6 

# Going through each employee row one by one
for i in range(len(final_df)):
    # Checking if the project status is 'Failed'
    if final_df.loc[i, 'Status'] == 'Failed':
        # Demoting the employee by increasing their designation level by 1
        final_df.loc[i, 'Designation Level'] = final_df.loc[i, 'Designation Level'] + 1

# Creating a new DataFrame that only includes employees with designation level 4 or below
final_df = final_df[final_df['Designation Level'] <= 4]

# Displaying the updated data
final_df
```

```python
# TASK 7 
# Defining a function to add titles based on gender
def add_title(row):
    if row['Gender'] == 'M':
        return 'Mr. ' + row['First Name']
    else:
        return 'Mrs. ' + row['First Name']

# Applying the function to each row to update the 'First Name' column
final_df['First Name'] = final_df.apply(add_title, axis=1)

# Removing the 'Gender' column as asked
final_df = final_df.drop('Gender', axis=1)

# Step 4: Displaying the updated DataFrame
final_df
```

```python
# TASK 8 

# Loop through each row in the DataFrame
for i in range(len(final_df)):
    # Checking if the employee's age is greater than 29
    if final_df['Age'].iloc[i] > 29:  # 'if' condition is used here
        # If the condition is true, promoting the employee by reducing their designation level by 1
        final_df.loc[i, 'Designation Level'] -= 1  # Using .loc to safely update the value

# Step 3: Showing the updated DataFrame
final_df
```

```python
# TASK 9 

# Grouping by ID and First Name to calculate total cost
total_proj_cost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Renaming the column to 'Total Cost'
total_proj_cost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Saving the new dataframe as 'TotalProjCost'
TotalProjCost = total_proj_cost

# Displaying the total project cost per employee
TotalProjCost
```

```python
# TASK 10

# Filtering employees whose city name contains the letter 'o'
filtered_employees = final_df[final_df['City'].str.contains('o', case=False)]

# Displaying the details of filtered employees
filtered_employees
```
